```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'
```

```python
params = { 'query_text' : 'codes'}
```

```python
response = requests.get(endpoint,params = params)
```

```python
code_search = response.json()
```

```python
df = pd.DataFrame(code_search)
```

```python
df['query_count']=df['query_count' ]. astype(int)
```

```python
over100 = df[df['query_count'] > 100]
```

```python
over100
```

```python
bymonth = over100.groupby(['month_name', 'year']).sum()
```
